- 1호선은 주요 터미널과 환승역을 통해 이용률을 유지, 2호선의 순환 효과에는 미치지 못함
- 4호선은 강남, 동대문 등 핵심 지역을 연결, 중복 효과가 적어 평균 승·하차 인원수가 상대적으로 낮게 나타남
- 3호선은 서울 외곽과 덜 밀집된 지역을 주로 연결함

```python
df2_st =  df[(df['호선명'] == '2호선') & (df['사용월'] == 202211)]
df2_st
```

```python
#승차인원 추출
df2_on = pd.DataFrame()
df2_on['지하철역'] = df2_st['지하철역']
for i in range(3, 50, 2):
    df2_on[df2_st.columns[i]] = df2_st[df2_st.columns[i]]
df2_on
```

```python
#하차인원 추출
df2_off = pd.DataFrame()
df2_off['지하철역'] = df2_st['지하철역']
for i in range(4, 51, 2):
    df2_off[df2_st.columns[i]] = df2_st[df2_st.columns[i]]
df2_off
```

```python
df2_on = df2_on.apply(pd.to_numeric, errors='coerce').fillna(0)
df2_off = df2_off.apply(pd.to_numeric, errors='coerce').fillna(0)
df = pd.DataFrame()
df['지하철역'] = df2_st['지하철역']
df['평균 승차 인원'] = df2_on.mean(axis=1).astype(int)
df['평균 하차 인원'] = df2_off.mean(axis=1).astype(int)
df = df.set_index('지하철역')
df_sortedon = df.sort_values('평균 승차 인원', ascending=False).head(10)
df_sortedon
```

#1.3 데이터 시각화

```python
plt.figure(figsize=[20,10])
plt.title('2호선 평균 승차 인원 수 상위 10', fontsize=20)
plt.bar(df_sortedon.index, df_sortedon['평균 승차 인원'])
```

- 2호선기준 잠실>강남>홍대입구>신림>구로디지털단지 순으로 평균 승차 인원이 많음
>>
- 잠실역이 주요 환승역 및 출퇴근 중심지 역할로 가장 많음
- 강남역은 상업 및 업무지구와의 연계로 높은 이용률을 기록
- 홍대입구역은 문화 및 유흥 중심지 인접성과 젊은층의 수요로 인해 승차 인원이 많음
- 신림역과 구로디지털단지는 각각 인구 밀집 지역과 산업 단지로서 일정한 승차 수요가 발생
